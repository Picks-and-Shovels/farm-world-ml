```python
import pandas as pd

from sklearn.metrics.pairwise import cosine_similarity
```

```python
userId = 0

board_data = pd.read_csv("data/board_read_like.csv")
farm_data = pd.read_csv("data/farm.csv")
board_tag = pd.read_csv("data/board_tag.csv")
```

```python
board_data.head()
```

```python
farm_data.head()
```

```python
df_user_board = board_data.pivot_table(values='viewed', index='userId', columns='boardId', aggfunc='sum').fillna(0)
```

```python
df_user_board.head()
```

```python
df_user_board = df_user_board.transpose()
df_user_board.head(10)
```

```python
sim = cosine_similarity(df_user_board, df_user_board)
print(sim.shape)
```

```python
sim_df = pd.DataFrame(data=sim, index=df_user_board.index, columns=df_user_board.index)
```

```python
sim_df.head()
```

```python
sim_df[userId].sort_values(ascending=False)[1:10]
```

```python
sim_view = []
sim_view.extend(sim_df[userId].sort_values(ascending=False)[1:11].index)
print(sim_view)
```

```python
farm_data = farm_data[['userId', 'crop0', 'crop1', 'crop2', 'crop3']]
crop_set = set(farm_data.loc[farm_data['userId'] == userId, ['crop0', 'crop1', 'crop2', 'crop3']].values.flatten())
crop_list = list(crop_set)
print(crop_list)
```

```python
board_tag.head()
```

```python
board_like = board_data[['boardId', 'liked']]
board_like_count = board_like.groupby('boardId')['liked'].sum().reset_index()
df_tag_like = pd.merge(board_tag, board_like_count, on='boardId')
df_tag_like.sort_values(by='liked', ascending=False, inplace=True)
df_tag_like.head(10)
```

```python
tag_crop = []

print(crop_list)
for i in crop_list:
	tag_crop.extend(df_tag_like.loc[df_tag_like['tagId'] == i, 'boardId'].values[:3])
print(tag_crop)
```

```python
# values : liked
sim_liked = []

df_user_board = board_data.pivot_table(
	values='liked',
	index='userId',
	columns='boardId',
	aggfunc='sum').fillna(0)

df_user_board = df_user_board.transpose()

sim = cosine_similarity(df_user_board, df_user_board)
sim_df = pd.DataFrame(data=sim, index=df_user_board.index, columns=df_user_board.index)

sim_liked.extend(sim_df[userId].sort_values(ascending=False)[1:11].index)
print(sim_liked)
```

```python
print(sim_view)
print(sim_liked)
print(tag_crop)
```

```python
same_numbers = set(sim_view) & set(sim_liked) & set(tag_crop)
print(same_numbers)
```
